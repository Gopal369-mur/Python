```python
#create and function that tells eligibility of marriage for male and female according to their age limit like 21 for male and 18 for female
class EligibilityforMarriage:
    def Eligible():
        gender = str(input("Enter the Gender:"))
        age = int(input("Enter the age:"))
        
        if gender == "male":
            if age >= 21:
                print("you are eligible for marriage")
            else:
                print("You are not eligible")
                
        elif gender == "female":
            if age >= 18:
                print(" you are eligible for marriage")
            else:
                print("You are not eligible")
EligibilityforMarriage.Eligible()
    
    
```
