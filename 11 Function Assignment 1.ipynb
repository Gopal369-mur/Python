```python
#Create a function that tells eligibility of marriage for male and female accoridng to their age limit like 21 for male and 18 for female
def Eligibilityformarriage():
    gender=str(input("Enter the Gender: "))
    age = int (input("Enter the age:"))
    if gender == "male":
        if age >=21:
            print("you are eligible for marriage")
        else:
            print("Not eligible")
    elif gender == "female":
        if age >= 18:
            print("you are eligible for marriage")
        else:
            print("Not eligible")
Eligibilityformarriage()
```
